```python
#4
n=input()
dangerous=False
one=0
zero=0
for i in range(0,len(n)):
        if(n[i] == '1'):
            one+=1
            zero = 0
        else:
            zero+=1
            one=0
        if(one == 7 or zero == 7):
            dangerous=True
            break
if dangerous==True:
    print("YES") 
else:
    print("NO")
```
